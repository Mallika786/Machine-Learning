```python
all_pixel=img.reshape(4012*6010,3)
all_pixel.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=19)
```

```python
km.fit(all_pixel)
```

```python
centre=km.cluster_centers_
centre
```

```python
centre=np.array(centre,dtype='uint')
centre
```

```python
km.labels_.shape
```

```python
np.unique(km.labels_,return_counts=True)
```

```python
new_img=np.zeros((all_pixel.shape[0],3),dtype='uint')
```

```python
img1.shape
```

```python
plt.imshow(new_img.reshape(4012,6010,3))
plt.show()
```
