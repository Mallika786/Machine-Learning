```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array(list(range(10)))
b=np.arange(0,10)
b
```

```python
plt.scatter(a,b)
plt.show()
```

```python
plt.scatter(a,b)
plt.grid()
plt.show()
```

```python
plt.scatter(a,b)
plt.grid()
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
plt.plot(a,b)
plt.plot(a,b**2)
plt.plot(a,b**3)
plt.grid()
plt.show()
```

```python
a=np.random.randint(0,10,10)
c=np.random.randint(0,10,10)
b=np.array(list(range(len(a))))
```

```python
plt.bar(b,a)
plt.bar(b+0.25,c,0.25)
plt.show()
```

```python
a=['a','b','c']
b=[10,20,40]
```

```python
plt.pie(b,labels=a)
plt.show()
```

```python
plt.pie(b,labels=a,explode=[0,0.25,0])
plt.show()
```
